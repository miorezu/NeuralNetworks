1. Виконати дії з векторами і матрицями:

a. Створити квадратну матрицю (n × n), заповнити її за правилом з таблиці
варіантів. Знайти детермінант матриці і обернену до неї матрицю.

b. Задати два вектори з n елементами, заповнені випадковими числами: один –
дійсними числами, другий – цілими числами. Обчислити їх суму і добуток.

c. Задати дві матриці, заповнені випадковими числами: одна розміру n × m, друга
– m × n. Знайти їх добуток.

d. Знайти добуток матриць і векторів (розбити на дві пари, для яких можливо
виконання операції).

```python
import numpy as np
```

```python
n = 3
m = 5
matrix = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        matrix[i][j] = pow((2 * i - j), 2 - j)
print("Our matrix:\n", matrix)
print("Det:", np.linalg.det(matrix))
print("Inverse:\n", np.linalg.inv(matrix))
```

```python
v1 = np.random.randn(n)
print("v1 = ", v1)
v2 = np.random.default_rng().integers(97, size = n)
print("v2 = ", v2)
print("Sum: ", v1+v2)
print("Multiplication: ", np.dot(v1, v2))
```

```python
M1 = np.random.randn(n,m)
M2 = np.random.randn(m,n)
print("M1:\n", M1)
print("M2:\n", M2)
print("Multiplication M1 * M2:\n", np.dot(M1, M2))
print("Multiplication M2 * M1:\n", np.dot(M2, M1))
```

```python
print("Multiplication v1 * M1:\n", np.dot(v1, M1))
print("Multiplication v2 * M1:\n", np.dot(v2, M1))
print("Multiplication M2 * v1:\n", np.dot(M2, v1))
print("Multiplication M2 * v2:\n", np.dot(M2, v2))
```
